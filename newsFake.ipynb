DATA PREPROCESSING

Preprocessing text. removeing extra space,special chr,url and links and covert to lowercase

```python
import re
import string
def wordopt(text):
    text = text.lower()  # convert to lowercase
    text = re.sub('\[.*?\]', '', text)  # remove text in brackets
    text = re.sub('https?://\S+|www\.\S+', '', text)  # remove URLs
    text = re.sub('<.*?>+', '', text)  # remove HTML tags
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)  # remove punctuation
    text = re.sub('\w*\d\w*', '', text)  # remove words containing numbers
    text = re.sub('\s+', ' ', text).strip()  # remove extra spaces
    text = re.sub("['\"“”‘’]", '', text)  # remove fancy/single/double quotes
    return text
```

```python
data["text"] = data["text"].apply(wordopt)
```

1. TOKENIZATION

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
data["tokens"] = data["text"].apply(lambda x: [token.text for token in nlp(x)])
#takes a lot of time
```

2. STOP WORD REMOVAL

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
def remove_stopwords(tokens):
    return [token for token in tokens if token.lower() not in STOP_WORDS]

data["tokens"] = data["tokens"].apply(remove_stopwords)
```

3. LEMMATIZATION

```python
def lemmatize_tokens(tokens):
    doc = nlp(" ".join(tokens))  #convert list of tokens to a string
    return [token.lemma_ for token in doc]

data["tokens"] = data["tokens"].apply(lemmatize_tokens)
```

5. Convert sentences to vectors:- word2vector gives word embeddings- but ML models need sentence/doc vectors.

```python
def sentence_vector(tokens, model):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    return np.mean(vectors,axis=0) if vectors else np.zeros(model.vector_size)
```

```python
data["vector"] = data["tokens"].apply(lambda x: sentence_vector(x,model))
```
